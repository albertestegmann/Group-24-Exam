```python
#Loading the merged csv to clean the last parts (se the other Jupyter files)
df_last = pd.read_csv("final_data_with_housing.csv")  
df_last
```

```python
#We now clean the monthly expenses where there was a mixup between python formating and the formating on Boligsiden
import pandas as pd

def clean_expense(value):
    value_str = str(value).strip() 
    
    # first we remove the python decimal suffix 
    if value_str.endswith('.000'):
        value_str = value_str[:-4] 
    elif value_str.endswith('.0') and len(value_str) > 2:
        value_str = value_str[:-2]  # Remove '.0'
    #Then we format a bit to make sure that the decimal place are the same for all values 
    try:
        number = float(value_str)
        
        if number.is_integer():
            value_str = str(int(number))  
        else:
            value_str = f"{number:.3f}" 
        
    except ValueError as e:
        print(f"Error processing value {value_str}: {e}")
        return None

    # then we add ',00' to every cleaned value to ensure consistency, as otherwise we would still get some misinterpretations and values would have different formats
    formatted_value = f"{value_str},00"
    
    # We then print the value after cleaning and formatting to ensure consistency 
    print(f"Cleaned and formatted value: {formatted_value}")
    
    return formatted_value


# Create a copy of df_cleaned2, so that if there are errors we don't have to redo that part 
df_last3 = df_last.copy()

# Convert the column to string type in the copied DataFrame
df_last3['Monthly expense'] = df_last3['Monthly expense'].astype(str)

# Apply the cleaning function to the 'Monthly expense' column in the new DataFrame
df_last3['Monthly expense'] = df_last3['Monthly expense'].apply(clean_expense)

# We then print the result again to ensure that everything is correct 
print(df_last3)
```

```python
#We now have a look at the property type column which is in some cases Villa and somehting else, 
#orginally this column was only included to make sure we didn't have apartments or other this by accident 

# Having a look at the houses that are Villa and something else 
filtered_df = df_last3[df_last3['Property type'].str.contains(r'^Villa/', na=False)]

# Display the filtered DataFrame
filtered_df.head(30)
```

```python
# We decide to create dummy variables for 'Landejendom' and 'Fritidsbolig' so that we can use them as a variable if we feel like it 
df_last3['Landejendom'] = df_last3['Property type'].str.contains('Landejendom').astype(int)
df_last3['Fritidsbolig'] = df_last3['Property type'].str.contains('Fritidsbolig').astype(int)
```

```python
#Now we make a new copy to make sure that we have a backup, just incase. 
df_last4 = df_last3.copy()

# Replace 'Villa/Landejendom' and 'Villa/Fritidsbolig' with just 'Villa'
df_last4['Property type'] = df_last4['Property type'].str.replace(r'/Landejendom|/Fritidsbolig', '', regex=True)
print(df_last4)
```

```python
# We then convert kr to Mio. kr so that we dont get the strange e+ something numbers as they are ugly to look at 
df_last4['Elderly Care expense (mio.)'] = df_last4['Elderly Care expense'] / 1e6
df_last4['Elementary school expense (mio.)'] = df_last4['Elementary school expense'] / 1e6
df_last4 = df_last4.drop(columns='Elementary school expense').drop(columns='Elderly Care expense')
#checking to see if everything looks nice and it does so that was lucky 
print(df_last4)

```

```python
#We now check some of the other columns to make sure that nothing is probematic 
df_subset = df_last4.iloc[:, 10:] 
print(df_subset)
```

```python
#There are some formating differences between columns that we now fix so that everything uses '.' as a decimal seperator 

df_last4['Off. vurdering'] = df_last4['Off. vurdering'].str.replace('.', '', regex=False).astype(float)
df_last4['Monthly expense'] = df_last4['Monthly expense'].str.replace('.', '', regex=False)  # Remove the period
df_last4['Monthly expense'] = df_last4['Monthly expense'].str.replace(',', '.', regex=False)  # Replace comma with period
df_last4['Monthly expense'] = df_last4['Monthly expense'].astype(float)  # Convert to float
```
